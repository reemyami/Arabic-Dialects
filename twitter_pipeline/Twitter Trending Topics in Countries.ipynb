```python
#importing all dependencies
import numpy as np
import tweepy
import requests
import base64
import json
from datetime import datetime    
import pandas as pd
```

```python
#Add your keys
consumer_key = ''
consumer_secret = ''
```

```python
#Reformat the keys and encode them
key_secret = '{}:{}'.format(consumer_key, consumer_secret).encode('ascii')
#Transform from bytes to bytes that can be printed
b64_encoded_key = base64.b64encode(key_secret)
#Transform from bytes back into Unicode
b64_encoded_key = b64_encoded_key.decode('ascii')
```

```python
base_url = 'https://api.twitter.com/'
auth_url = '{}oauth2/token'.format(base_url)
```

```python
auth_headers = {
    'Authorization': 'Basic {}'.format(b64_encoded_key),
    'Content-Type': 'application/x-www-form-urlencoded;charset=UTF-8'
}
auth_data = {
    'grant_type': 'client_credentials'
}


auth_resp = requests.post(auth_url, headers=auth_headers, data=auth_data)
```

```python
access_token = auth_resp.json()['access_token']
print(auth_resp.status_code)
trend_headers = {
    'Authorization': 'Bearer {}'.format(access_token)    
}
```

```python
#country and their corresponding WOEID
countries = {
    'Saudi': 23424938, 
    'Eygpt': 23424802, 
    'Kuwait': 23424870, 
    'Oman':23424898, 
    'Lebanon': 23424873, 
    'Jordan': 23424860, 
    'Algeria':23424740}
```

```python
#return country name
def country_name(id):
    if id == countries['Saudi']:
        return 'Saudi'
    elif id == countries['Eygpt']:
         return 'Eygpt'
    elif id == countries['Algeria']:
        return 'Algeria'
    elif id == countries['Lebanon']:
        return 'Lebanon'
    elif id == countries['Jordan']:
        return 'Jordan'
    elif id == countries['Oman']:
        return 'Oman'
    elif id == countries['Kuwait']:
        return 'Kuwait'
```

```python
#Return the date to utilise as a file name
datestring = datetime.strftime(datetime.now(), '%Y_%m_%d')
now = datetime.now()
```

```python
trend_url = 'https://api.twitter.com/1.1/trends/place.json'

#loop through the countries and return trends
for key in countries:
    
    print('\n',key,countries[key])
    
    trend_params = {'id' : countries[key],}
    #print(trend_params)
    
    trend_resp = requests.get(trend_url, headers=trend_headers, params=trend_params)
    #print(trend_resp)
    
    tweet_data = trend_resp.json()
    #print(tweet_data)
    
    trend = []
    length = len(tweet_data[0]['trends'])
    if length == 0: 
        print("list is empty")
    else:
        print('#Trends Retrieved from', key, 'is : ', length)
    for i in range(0,length):
        trend.append(tweet_data[0]['trends'][i]['name'])

    
    #Print Trends
    print(*trend , sep='\n')
    
    
    # dictionary  
    dict = {'trend': trend}  
    print('Structuring Trends...')     
    df = pd.DataFrame(dict) 
    country= country_name(trend_params['id'])
    
    path = country+'_trends_'+datestring+'.csv'
    print('Saving to...',path)
    df.to_csv(path)
    print(key, 'Trends are Saved to File..')
```
